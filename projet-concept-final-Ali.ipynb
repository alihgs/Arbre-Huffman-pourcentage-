```python
import tkinter as tk


class ArbreB(object):
    def __init__(self, etiquette, nom, f_gauche=None ,f_droite=None):
        self.nom=nom
        self.etiquette = etiquette
        self.f_gauche=f_gauche
        self.f_droite=f_droite
        

    def lister(self):
        ''' retourne une liste de feuilles '''
        liste=[]
        if self.nom==None:
            self.f_gauche.lister()
            self.f_droite.lister()
        elif self.nom!=None:
            liste.append(ArbreB(self.etiquette,self.nom))
        return liste
 
     
class Sommet(object):
    def __init__(self,etiquette=None, f_gauche=None, f_droite=None, nom=None):
        self.etiquette = etiquette
        self.f_gauche = f_gauche
        self.f_droite = f_droite
        self.nom=nom
    

    def lister(self):
        
        if self.nom==None:
            return []+self.f_gauche.lister()+self.f_droite.lister()

            
    def inserer(self, A):
        l=self.lister()
        l.append(A)
        self.fusion(l)

    def supprimer(self,N):
        l=self.lister()
        for elem in l:
            if elem.nom==N:
                l.pop(l.index(elem))
        self.fusion(l)


    def fusion(self,l):
        '''la méthode fusion permet de créer un arbre de Huffman en combinant de manière répétée
          les deux nœuds les plus légers jusqu'à ce qu'il ne reste plus qu'un seul nœud,
          qui représente la racine de l'arbre de Huffman.'''
        while len(l)>1:
            l.sort( key=lambda poid: poid.etiquette)
            S=Sommet(l[0].etiquette+l[1].etiquette,l[0],l[1])
            l.append(S)
            l.pop(0)        
            l.pop(0)
        self.etiquette = l[0].etiquette
        self.f_gauche = l[0].f_gauche
        self.f_droite = l[0].f_droite

#Partie 2
#1


def nbrO(txt):
     return {lettre : (txt.count(lettre)*100)//len(txt) for lettre in txt }


#2
def creation_list_arbre(dic):
    return [ArbreB(val, lettre) for lettre,val in dic.items()]

#3
def codage(so,tt="",d={}):
    
    if so.nom==None:
        if so.f_gauche!=None:
            tt=tt+"0"
            d=codage(so.f_gauche,tt)
            tt=tt[:-1]
        if so.f_droite!=None:
            tt=tt+"1"
            d=codage(so.f_droite,tt)
            tt=tt[:-1]
    else:
        d[so.nom]=tt
    return d
def coddage_txt(t,d):
    chaine=""
    for lettre in t:
        chaine+=d[lettre]
    return chaine
#partie 3
def decodage(t,d):
    x=""
    txt1=""
    for lettre in t:
        x=x+lettre
        for key,values in d.items():
            if x==values:
                txt1=txt1+key
                x=""

    return str(txt1)
        


    




T= "llourd"
dic=nbrO(T)
dic=dict(sorted(dic.items(), key=lambda t: t[1]))
print(dic)
l=creation_list_arbre(dic)
s = Sommet()
s.fusion(l)
diccodage=codage(s)
print(diccodage)

print(T)
txtcode=coddage_txt(T,diccodage)
print(txtcode)
print(decodage(txtcode,diccodage))

```

```python
                                                   #"""Partie interface graphique"""

class interface(tk.Frame):
    def __init__(self, fenetreParent):
        '''Initialisation du canvas en heritant de la class tk.Frame'''
        tk.Frame.__init__(self, fenetreParent)
        self.parent = fenetreParent
        self.canvas = tk.Canvas(self, width=1000, height=800)
        self.canvas.pack(fill=tk.BOTH, expand=True)
        

    def dessinAbr(self, sommet, x, y, r, delta_y):
        if sommet is None:
            return

        # Dessiner le cercle avec texte en dessous
        self.canvas.create_oval(x-r, y-r, x+r, y+r, fill='white')
        if sommet.nom != None:
            self.canvas.create_text(x, y+r+10, text={sommet.nom: sommet.etiquette})
        else:
            self.canvas.create_text(x, y+r+10, text=sommet.etiquette)

        # Dessiner les sous-arbres gauches et droits
        if sommet.f_gauche != None:
            # Calculer les coordonnées du centre du sous-arbre gauche
            x_gauche = x - 2*r
            y_gauche = y + delta_y
            self.canvas.create_line(x-r*(2)**(-1/2), y+r*(2)**(-1/2), x_gauche, y_gauche, width=2)
            self.dessinAbr(sommet.f_gauche, x_gauche, y_gauche, r/2, delta_y)

        if sommet.f_droite != None:
            # Calculer les coordonnées du centre du sous-arbre droit
            x_droit = x + 2*r
            y_droit = y + delta_y
            self.canvas.create_line(x+r*(2)**(-1/2), y+r*(2)**(-1/2), x_droit, y_droit, width=2)
            self.dessinAbr(sommet.f_droite, x_droit, y_droit, r/2, delta_y)

    def positionRacine(self, sommet):
        # Dessiner l'arbre avec la racine au milieu
        x = 500
        y = 110
        r = 100
        delta_y = r*3/2
        self.dessinAbr(sommet, x, y, r, delta_y)


# Créer l'interface graphique principale
fenetre = tk.Tk()

txtP=""
frequenceP=""
listeP=[]
dic_codageP={}
codeTxtP=""
decodeTxtP=""
#Fonction à utiliser sur un bouton pour afficher l'arbre
def afficheTxt():
    global txtP
    global frequenceP
    global listeP
    global dic_codageP
    global codeTxtP


    txtP = zone_txt.get()

    frequenceP = nbrO(txtP)
    listeP = creation_list_arbre(frequenceP)
    s = Sommet()
    s.fusion(listeP)

    #txt code
    dico_codageP = codage(s)
    codeTxtP = coddage_txt(txtP,dico_codageP)
    affichage_codeTxt.config(text=f"Code du texte: {codeTxtP}", fg="red")

    fenetre2 = tk.Tk()
    fenetre2.title("Arbre de Huffman")
    canva = interface(fenetre2)
    canva.pack(fill=tk.BOTH, expand=True)
    canva.positionRacine(s)


#Fonction à utiliser sur un bouton pour decoder le texte
def decodageTxt():
    global txtP
    global frequenceP
    global listeP
    global dic_codageP
    global codeTxtP
    global decodeTxtP

    txtP = zone_txt.get()

    frequenceP = nbrO(txtP)
    listeP = creation_list_arbre(frequenceP)
    s = Sommet()
    s.fusion(listeP)

    dico_codageP = codage(s)
    codeTxtP = coddage_txt(txtP,dico_codageP)

    decodeTxtP = decodage(codeTxtP,dico_codageP)
    affichage_decodeTxt.config(text=f"Decodage de Huffman: {decodeTxtP}", fg="blue")

#Fonction à utiliser sur un bouton pour effacer le texte entré
def effacerTxt():
    zone_txt.delete(0, 'end')
    affichage_codeTxt.config(text="", fg="red")
    affichage_decodeTxt.config(text="", fg="blue")



#Zone commentaire fenetre principale
entree_label_1 = tk.Label(fenetre, text="Veillez entrer votre texte ci-desous")
entree_label_1.pack(pady=60)

entree_label_2 = tk.Label(fenetre, text="Cliquez sur le bouton pour afficher l'arbre de Huffman correspondant à votre texte")
entree_label_2.pack()

entree_label_3 = tk.Label(fenetre, text="Avant de créer un nouvel arbre, veuillez d'abord effacer votre texte")
entree_label_3.pack()

#Zone de txt
zone_txt = tk.Entry(fenetre)
zone_txt.pack()

    #Commentaire codage
affichage_codeTxt = tk.Label(fenetre, text=None, height=5, width=50)
affichage_codeTxt.pack(fill=tk.BOTH, expand=True)

    #Commentaire decodage
affichage_decodeTxt = tk.Label(fenetre, text=None, height=5, width=50)
affichage_decodeTxt.pack(fill=tk.BOTH, expand=True)



#Boutton afficher txt
bouton_getTxt = tk.Button(fenetre, text="Affichage de l'arbre de Huffman", command=afficheTxt)
bouton_getTxt.pack()

#
bouton_decodage = tk.Button(fenetre, text="Decodage du texte", command=decodageTxt)
bouton_decodage.pack()

#Bouton effacer txt
bouton_removeTxt = bouton_getTxt = tk.Button(fenetre, text="Effacer le texte", command=effacerTxt)
bouton_getTxt.pack()

# # Affichage de la fenetre
canva = interface(fenetre)
canva.pack(fill=tk.BOTH, expand=True)


# Afficher l'interface graphique
canva.mainloop()
```
